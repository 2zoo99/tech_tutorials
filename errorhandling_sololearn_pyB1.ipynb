## error handling  
컴퓨터는 위에서 부터 아래로 한 줄 한 줄 실행지침에 맞게 작성됐는지 검사한다.  
만약에, 첫 번째 오류가 발견된다면 프로그램은 즉시 실행을 중단한다.    
파이썬에서 오류는 `Bugs` 하고 `exceptions` 로 나뉜다.
- `Bugs` 는 프로그램 코드 안에 뭐가 빠졌거나, 로직을 잘못 입력한 경우에 나타나는 오류다.  
이 오류는 프로그램 실행이 중단되지는 않는다.  
하지만, 원하는 결과가 나오지 않거나, 원하는 동작으로 실행되지 않을 수 있다.
- `Exceptions` 는 프로그램 코드 안에 컴파일 지침에 맞지 않은 코드를 작성해서 실행이 되지 못하는 상황에 이르는 오류다.  
이 떄 프로그램 실행이 중단되고, 오류정정을 위해 예외사항에 대한 설명을 출력한다.  
파이썬 언어의 예외사항으로 중단되는 경우는 다음과 같이 있다.  
  - `NameError : variable is noe defined`  
  : 정의하지 않은 변수를 작성한 경우  
  = 즉, 변수 이름을 잘못 작성했을 때    
  - `SyntaxError : incorrect syntax`  
  : 프로그래밍 언어의 문법지침에 따르지 않은 경우  
  = 즉, 함수를 잘못 사용했거나, 연산자를 잘못사용했거나, 함수를 작성할때 특정기호를 빠트렸거나, 콤마를 안적었거나, 괄호를 안닫았거나, 콜론(:)을 안적었거나, ...etc
  - `IndexError : out of range index`   
  : 반복자 변수의 사용지침에 어긋난 경우  
  = 즉, 리스트 변수의 크기보다 큰 인덱스값에 접근하도록 적었을때, 인덱스 범위를 틀리게 적었을때
  - `TypeError : incorrect type of argument`  
  : 함수를 호출할때 함수에 들어가는 변수타입이 적합하지 않은 경우  
  = 즉, 문자를 받는 함수인데 정수를 집어넣었거나, 리스트 함수인데 단일 정수변수를 넣었거나, ... etc
  - `ValueError : inappropriate value`  
  : 함수를 호출할때 함수에 들어가는 변수타입은 맞지만, 변수값이 함수를 적용하기에 적절하지 않은 경우
  = 예를들면, 문자열을 정수로 변환하는 int() 함수를 사용한다고 할때 phone 변수가 '010-2222-3333' 일 경우 문자가 포함되어 있기 떄문에 int() 함수를 적용할 수 없음

### exceptions 는 가끔 예측이 가능할 때가 있다.
예외사항을 직접 다루거나, 프로그램 중단을 방지하고자 할때, 파이썬에서는 `try 와 except` 표현식을 사용한다.
  
- `try` 블록에서는 예외가 일어날 수도 있는 코드 부분을 작성해둔다.  
예외가 일어나지 않을 경우에는, 해당 `try` 에 적힌 코드가 실행된다.
- `except` 블록에는 `try` 블록에서 특정한 예외가 일어난다면 중단되지 않고 실행할 코드를 작성해둔다.  
전체 프로그램에서 예외가 일어나더라도 실행중단을 막는 역할을 할 수 있다.  
`except` 는 여러개 작성할 수 있다. e.g.
  - except:
  - except TypeError:
  - except IndexError:

`try/except` 표현식을 사용하면 실행 중단을 막을 수 있고, 원하는 동작을 자유자재로 통제하는 기능을 추가적으로 구현해볼 수 있다.

만약 try, except 둘 중에 실행되는 코드 상관없이 무조건 실행되게 하고 싶은 코드가 있다면 `finally` 표현식을 사용하면 된다.

(sololearn 출처)

```python
prices = [559, 879, "N/A", 349]
# prices = [559, 879, 349]

try:
  print(sum(prices))
except TypeError:
  print("Check the prices")
finally:
  print("Need help? Contact us")
```

아니면, try 코드가 실행되고 나서만 실행되는 특정 코드를 구현하고 싶다면 `else` 표현식을 사용하면 된다.
image.png
(sololearn 출처)

```python
books = ['harry potter', 'dune', 'emma']

try:
  choice = books[1]
except IndenxError:
  print("out of range")
else:
  print(choice + " is a great choice!")
```
